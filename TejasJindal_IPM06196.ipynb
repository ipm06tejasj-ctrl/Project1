## Load and Preprocess Data

### Subtask:
Load the 'household_power_consumption.txt' dataset. Clean the data by handling missing values, converting data types (especially date and time columns), and potentially resampling the data to a consistent frequency suitable for time-series analysis.


**Reasoning**:
The first step is to import the pandas library and load the 'household_power_consumption.txt' dataset into a DataFrame. I will inspect the first few rows and the data types to understand the structure and identify potential issues, paying attention to the separator.



```python
import pandas as pd

# Load the dataset, assuming ';' as a separator based on common datasets of this type.
df = pd.read_csv('/content/household_power_consumption.txt', sep=';', low_memory=False)

# Display the first few rows of the DataFrame
print('First 5 rows of the dataset:')
print(df.head())

# Display the data types of each column
print('\nData types of each column:')
print(df.info())
```

**Reasoning**:
The previous step showed that several columns are of 'object' type and likely contain non-numeric values (like '?') representing missing data. Before converting these to numeric types and combining date/time, I need to identify and handle these missing values. I'll replace '?' with NaN and then convert the relevant columns to numeric types, coercing errors to NaN.



```python
import numpy as np

# Replace '?' with NaN across the DataFrame
df.replace('?', np.nan, inplace=True)

# Convert relevant columns to numeric
# These are the columns identified as object type that should be numeric
numeric_cols = [
    'Global_active_power', 'Global_reactive_power', 'Voltage',
    'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check for missing values after conversion
print('Missing values after converting to numeric:')
print(df.isnull().sum())

# Display the updated data types
print('\nUpdated data types of each column:')
print(df.info())
```

**Reasoning**:
Now that the numeric columns have been converted and potential missing values handled, the next step is to combine the 'Date' and 'Time' columns into a single datetime object. This new datetime column will then be set as the DataFrame's index, and the DataFrame will be sorted to ensure all entries are in chronological order for time-series analysis.



```python
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')
df.set_index('DateTime', inplace=True)
df.sort_index(inplace=True)

# Drop the original 'Date' and 'Time' columns as they are now redundant
df.drop(columns=['Date', 'Time'], inplace=True)

print('DataFrame after combining Date and Time and setting index:')
print(df.head())

print('\nUpdated data types and index information:')
print(df.info())
```

**Reasoning**:
The previous steps have handled missing values, converted data types, and set the DateTime index. Now, I need to address the remaining missing values in the numeric columns before resampling. I will impute these few remaining NaN values using forward fill, which is a common practice in time series data to carry forward the last valid observation. After imputation, I will resample the data to an hourly frequency, aggregating the numerical columns by their mean, as the original data seems to be at a minute-level frequency and an hourly frequency is often more appropriate for this type of analysis.



```python
import pandas as pd

# Impute remaining missing values using forward fill
df.fillna(method='ffill', inplace=True)

# Check for missing values after imputation
print('Missing values after forward fill imputation:')
print(df.isnull().sum())

# Resample the data to hourly frequency, taking the mean of numerical columns
df_hourly = df.resample('H').mean()

print('\nDataFrame after hourly resampling:')
print(df_hourly.head())

print('\nUpdated data types and index information for hourly data:')
print(df_hourly.info())
```

**Reasoning**:
The previous code block produced `FutureWarning` messages regarding the usage of `df.fillna(method='ffill')` and `df.resample('H')`. To align with future Pandas versions and suppress these warnings, I will update the code to use `df.ffill()` directly and replace the deprecated frequency alias 'H' with 'h'.



```python
import pandas as pd

# Impute remaining missing values using forward fill
df.ffill(inplace=True)

# Check for missing values after imputation
print('Missing values after forward fill imputation:')
print(df.isnull().sum())

# Resample the data to hourly frequency, taking the mean of numerical columns
df_hourly = df.resample('h').mean()

print('\nDataFrame after hourly resampling:')
print(df_hourly.head())

print('\nUpdated data types and index information for hourly data:')
print(df_hourly.info())
```

## Feature Engineering for Clustering

### Subtask:
Extract daily features from the hourly resampled data, such as average, maximum, and standard deviation of 'Global_active_power', to characterize daily consumption patterns for subsequent clustering.


**Reasoning**:
The subtask requires extracting daily features from the hourly resampled data. I will extract the date from the index, group the data by day, and then calculate the mean, maximum, and standard deviation of 'Global_active_power' for each day, storing these in a new DataFrame called `df_daily_features`.



```python
df_daily_features = df_hourly.groupby(df_hourly.index.date)['Global_active_power'].agg(
    daily_avg_power='mean',
    daily_max_power='max',
    daily_std_power='std'
)

# Convert the index back to datetime objects if needed, or keep as date objects
df_daily_features.index = pd.to_datetime(df_daily_features.index)

print('Daily features DataFrame head:')
print(df_daily_features.head())

print('\nDaily features DataFrame info:')
print(df_daily_features.info())
```

## Apply Unsupervised Clustering

### Subtask:
Apply K-Means clustering to group households based on their consumption behavior using the engineered features. Determine an appropriate number of clusters, potentially using methods like the elbow method or silhouette score. Visualize the distribution of clusters.


**Reasoning**:
The next step is to prepare the data for clustering by scaling the features using StandardScaler and then apply the Elbow Method to determine an optimal number of clusters for K-Means.



```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 2. Initialize a StandardScaler and fit-transform df_daily_features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_daily_features)

# Convert scaled_features back to a DataFrame for easier handling if needed later, but not strictly necessary for KMeans
scaled_df = pd.DataFrame(scaled_features, columns=df_daily_features.columns, index=df_daily_features.index)

# 3. Implement the Elbow Method to find an optimal number of clusters
inertia = []
range_n_clusters = range(1, 11) # Test from 1 to 10 clusters

for i in range_n_clusters:
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10) # Set n_init to avoid warning
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Plot the inertia values against the number of clusters
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()

print("Features scaled and Elbow method plot displayed. Next, determine n_clusters from the plot.")
```

**Reasoning**:
Based on the elbow plot from the previous step, an optimal number of clusters will be chosen. Then, K-Means clustering will be applied with this chosen number of clusters to assign labels to each daily feature. Finally, PCA will be used to reduce the dimensionality for visualization, and the clusters will be plotted to show their distribution.



```python
n_clusters = 3 # Chosen based on the elbow plot where the decrease in inertia starts to slow down significantly

# 5. Initialize a KMeans model with the chosen n_clusters and fit it to scaled_features
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(scaled_features)

# 6. Add the cluster labels to df_daily_features as a new column
df_daily_features['cluster_label'] = cluster_labels

# 7. Apply Principal Component Analysis (PCA) for visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)

# Convert pca_components to a DataFrame for easier plotting and to retain index
pca_df = pd.DataFrame(data = pca_components, columns = ['principal_component_1', 'principal_component_2'], index=df_daily_features.index)

# Add cluster labels to the PCA dataframe for coloring
pca_df['cluster_label'] = df_daily_features['cluster_label']

# 8. Create a scatter plot of the pca_components, colored by cluster_label
plt.figure(figsize=(12, 8))
for cluster_id in sorted(pca_df['cluster_label'].unique()):
    subset = pca_df[pca_df['cluster_label'] == cluster_id]
    plt.scatter(
        subset['principal_component_1'],
        subset['principal_component_2'],
        label=f'Cluster {cluster_id}',
        alpha=0.7
    )

plt.title('Clusters of Daily Power Consumption (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()

print(f"K-Means clustering applied with {n_clusters} clusters. Clusters visualized using PCA.")
```

## Prepare Data for Supervised Learning

### Subtask:
Prepare the dataset for supervised learning by creating lagged features or other time-series specific features that could help in forecasting. Split the data into training and testing sets, ensuring proper handling of time-series splits.


**Reasoning**:
The first step in preparing the dataset for supervised learning is to create lagged features, which are essential for time-series forecasting. I will create a one-hour lagged version of 'Global_active_power' and then handle the NaN values introduced by this operation. After that, I will define the feature set (X) including the lagged feature and the target variable (y).



```python
import pandas as pd

# 1. Create a new column in df_hourly named Global_active_power_lag1
df_hourly['Global_active_power_lag1'] = df_hourly['Global_active_power'].shift(1)

# 2. Drop any rows that contain NaN values resulting from the creation of lagged features.
df_hourly_cleaned = df_hourly.dropna()

# 3. Define the features (X) for the model.
# Include the lagged power and other relevant features that could be predictive.
X = df_hourly_cleaned[[
    'Global_active_power_lag1',
    'Global_reactive_power',
    'Voltage',
    'Global_intensity',
    'Sub_metering_1',
    'Sub_metering_2',
    'Sub_metering_3'
]]

# 4. Define the target variable (y) as the Global_active_power column from df_hourly.
y = df_hourly_cleaned['Global_active_power']

print('Features (X) head:')
print(X.head())

print('\nTarget (y) head:')
print(y.head())

print('\nShape of X:', X.shape)
print('Shape of y:', y.shape)
```

**Reasoning**:
The next step is to split the data into training and testing sets. For time series data, it's crucial to perform a chronological split to avoid data leakage. I will use a specified cut-off date to divide the data into training and testing portions for both features (X) and the target variable (y).



```python
cut_off_date = '2007-04-01'

# Split data into training and testing sets chronologically
X_train = X[X.index < cut_off_date]
X_test = X[X.index >= cut_off_date]
y_train = y[y.index < cut_off_date]
y_test = y[y.index >= cut_off_date]

print(f'Shape of X_train: {X_train.shape}')
print(f'Shape of X_test: {X_test.shape}')
print(f'Shape of y_train: {y_train.shape}')
print(f'Shape of y_test: {y_test.shape}')

print('\nData splitting complete. The dataset is now ready for supervised learning models.')
```

## Train Supervised Models (Global)

### Subtask:
Train a Random Forest Regressor and a Gradient Boosting Regressor on the entire dataset (without considering clusters initially) to serve as a baseline model. Make predictions on the test set.


**Reasoning**:
I need to import the necessary libraries for Random Forest and Gradient Boosting regressors, as well as the evaluation metrics. Then, I will initialize, train, make predictions, and evaluate both models on the test set as per the instructions.



```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# 3. Initialize a RandomForestRegressor model
rf_global_model = RandomForestRegressor(random_state=42)

# 4. Initialize a GradientBoostingRegressor model
gbr_global_model = GradientBoostingRegressor(random_state=42)

print('Training Random Forest Regressor...')
# 5. Train the rf_global_model
rf_global_model.fit(X_train, y_train)
print('Random Forest Regressor trained.')

print('Training Gradient Boosting Regressor...')
# 6. Train the gbr_global_model
gbr_global_model.fit(X_train, y_train)
print('Gradient Boosting Regressor trained.')

# 7. Make predictions on X_test using rf_global_model
rf_global_predictions = rf_global_model.predict(X_test)

# 8. Make predictions on X_test using gbr_global_model
gbr_global_predictions = gbr_global_model.predict(X_test)

# 9. Calculate evaluation metrics for Random Forest
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_global_predictions))
rf_mae = mean_absolute_error(y_test, rf_global_predictions)
rf_r2 = r2_score(y_test, rf_global_predictions)

# 10. Calculate evaluation metrics for Gradient Boosting
gbr_rmse = np.sqrt(mean_squared_error(y_test, gbr_global_predictions))
gbr_mae = mean_absolute_error(y_test, gbr_global_predictions)
gbr_r2 = r2_score(y_test, gbr_global_predictions)

# 11. Print the evaluation metrics
print('\n--- Global Model Evaluation Metrics ---')
print('\nRandom Forest Regressor:')
print(f'  RMSE: {rf_rmse:.4f}')
print(f'  MAE: {rf_mae:.4f}')
print(f'  R-squared: {rf_r2:.4f}')

print('\nGradient Boosting Regressor:')
print(f'  RMSE: {gbr_rmse:.4f}')
print(f'  MAE: {gbr_mae:.4f}')
print(f'  R-squared: {gbr_r2:.4f}')
```

## Train Supervised Models (Cluster-Based)

### Subtask:
For each identified cluster, train separate Random Forest Regressor and Gradient Boosting Regressor models. Make predictions for each cluster's test data.


**Reasoning**:
The first step is to integrate the `cluster_label` from the daily features back into the hourly cleaned DataFrame. This is crucial for splitting the data into clusters for subsequent model training. I will achieve this by creating a new column in `df_hourly_cleaned` and populating it by mapping the `cluster_label` based on the date of each hourly entry.



```python
import pandas as pd

# Map the cluster_label from df_daily_features to df_hourly_cleaned
# First, ensure the index of df_hourly_cleaned is datetime to extract date component correctly.
# df_hourly_cleaned.index is already DatetimeIndex

# Create a date column in df_hourly_cleaned to facilitate the merge/map
df_hourly_cleaned['date'] = df_hourly_cleaned.index.date

# Create a DataFrame for daily cluster labels with date as index
df_daily_cluster_map = df_daily_features[['cluster_label']].copy()
df_daily_cluster_map.index = df_daily_cluster_map.index.date

# Map the cluster labels
df_hourly_cleaned['cluster_label'] = df_hourly_cleaned['date'].map(df_daily_cluster_map['cluster_label'])

# Drop the temporary 'date' column
df_hourly_cleaned.drop(columns=['date'], inplace=True)

print('df_hourly_cleaned head with cluster_label:')
print(df_hourly_cleaned.head())

print('\nMissing cluster labels (should be 0 if mapping was successful):')
print(df_hourly_cleaned['cluster_label'].isnull().sum())

print('\nUnique cluster labels in df_hourly_cleaned:')
print(df_hourly_cleaned['cluster_label'].unique())
```

**Reasoning**:
The previous cell successfully added the `cluster_label` to `df_hourly_cleaned`, despite a `SettingWithCopyWarning`. Now, I will proceed with training cluster-based supervised models as outlined in the subtask instructions. This involves initializing dictionaries to store results, iterating through each cluster to filter data, splitting it into train/test sets, and then training and evaluating separate Random Forest and Gradient Boosting models for each cluster.



```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize empty dictionaries to store predictions and evaluation metrics for each cluster
rf_cluster_predictions = {}
gbr_cluster_predictions = {}
rf_cluster_metrics = {}
gbr_cluster_metrics = {}

# Define feature columns for X_cluster based on previous steps
feature_cols = [
    'Global_active_power_lag1', 'Global_reactive_power', 'Voltage',
    'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3'
]
target_col = 'Global_active_power'

# Define the cut-off date for train-test split, as used previously
cut_off_date = '2007-04-01'

# Iterate through each unique cluster ID
unique_clusters = sorted(df_hourly_cleaned['cluster_label'].unique())

for cluster_id in unique_clusters:
    print(f"\n--- Processing Cluster {cluster_id} ---")

    # Filter df_hourly_cleaned for the current cluster
    df_cluster = df_hourly_cleaned[df_hourly_cleaned['cluster_label'] == cluster_id].copy()

    # Define X_cluster (features) and y_cluster (target)
    X_cluster = df_cluster[feature_cols]
    y_cluster = df_cluster[target_col]

    # Split X_cluster and y_cluster into training and testing sets chronologically
    X_train_cluster = X_cluster[X_cluster.index < cut_off_date]
    X_test_cluster = X_cluster[X_cluster.index >= cut_off_date]
    y_train_cluster = y_cluster[y_cluster.index < cut_off_date]
    y_test_cluster = y_cluster[y_cluster.index >= cut_off_date]

    # Ensure there's enough data to train and test for the current cluster
    if X_train_cluster.empty or X_test_cluster.empty:
        print(f"Skipping cluster {cluster_id}: Insufficient data for training or testing after split.")
        continue

    # --- Random Forest Regressor ---
    # Initialize and train RandomForestRegressor
    rf_cluster_model = RandomForestRegressor(random_state=42)
    rf_cluster_model.fit(X_train_cluster, y_train_cluster)

    # Make predictions
    rf_predictions = rf_cluster_model.predict(X_test_cluster)
    rf_cluster_predictions[cluster_id] = rf_predictions

    # Calculate evaluation metrics for Random Forest
    rf_rmse = np.sqrt(mean_squared_error(y_test_cluster, rf_predictions))
    rf_mae = mean_absolute_error(y_test_cluster, rf_predictions)
    rf_r2 = r2_score(y_test_cluster, rf_predictions)
    rf_cluster_metrics[cluster_id] = {'RMSE': rf_rmse, 'MAE': rf_mae, 'R-squared': rf_r2}

    print(f"  Random Forest Regressor (Cluster {cluster_id}):")
    print(f"    RMSE: {rf_rmse:.4f}")
    print(f"    MAE: {rf_mae:.4f}")
    print(f"    R-squared: {rf_r2:.4f}")

    # --- Gradient Boosting Regressor ---
    # Initialize and train GradientBoostingRegressor
    gbr_cluster_model = GradientBoostingRegressor(random_state=42)
    gbr_cluster_model.fit(X_train_cluster, y_train_cluster)

    # Make predictions
    gbr_predictions = gbr_cluster_model.predict(X_test_cluster)
    gbr_cluster_predictions[cluster_id] = gbr_predictions

    # Calculate evaluation metrics for Gradient Boosting
    gbr_rmse = np.sqrt(mean_squared_error(y_test_cluster, gbr_predictions))
    gbr_mae = mean_absolute_error(y_test_cluster, gbr_predictions)
    gbr_r2 = r2_score(y_test_cluster, gbr_predictions)
    gbr_cluster_metrics[cluster_id] = {'RMSE': gbr_rmse, 'MAE': gbr_mae, 'R-squared': gbr_r2}

    print(f"  Gradient Boosting Regressor (Cluster {cluster_id}):")
    print(f"    RMSE: {gbr_rmse:.4f}")
    print(f"    MAE: {gbr_mae:.4f}")
    print(f"    R-squared: {gbr_r2:.4f}")

# Print all collected evaluation metrics after the loop
print("\n--- Cluster-Based Model Evaluation Metrics ---")
print("\nRandom Forest Regressor (by Cluster):")
for cluster_id, metrics in rf_cluster_metrics.items():
    print(f"  Cluster {cluster_id}:")
    print(f"    RMSE: {metrics['RMSE']:.4f}")
    print(f"    MAE: {metrics['MAE']:.4f}")
    print(f"    R-squared: {metrics['R-squared']:.4f}")

print("\nGradient Boosting Regressor (by Cluster):")
for cluster_id, metrics in gbr_cluster_metrics.items():
    print(f"  Cluster {cluster_id}:")
    print(f"    RMSE: {metrics['RMSE']:.4f}")
    print(f"    MAE: {metrics['MAE']:.4f}")
    print(f"    R-squared: {metrics['R-squared']:.4f}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 2. Create Bar Charts for Evaluation Metrics

# Consolidate metrics into a DataFrame
metrics_data = []

# Global Model Metrics
metrics_data.append({
    'Model': 'RF Global',
    'Type': 'Global',
    'RMSE': rf_rmse, # These are from the previous global model evaluation
    'MAE': rf_mae,
    'R-squared': rf_r2
})
metrics_data.append({
    'Model': 'GBR Global',
    'Type': 'Global',
    'RMSE': gbr_rmse, # These are from the previous global model evaluation
    'MAE': gbr_mae,
    'R-squared': gbr_r2
})

# Cluster-Based Model Metrics
for cluster_id, metrics in rf_cluster_metrics.items():
    metrics_data.append({
        'Model': f'RF Cluster {cluster_id}',
        'Type': f'Cluster {cluster_id}',
        'RMSE': metrics['RMSE'],
        'MAE': metrics['MAE'],
        'R-squared': metrics['R-squared']
    })

for cluster_id, metrics in gbr_cluster_metrics.items():
    metrics_data.append({
        'Model': f'GBR Cluster {cluster_id}',
        'Type': f'Cluster {cluster_id}',
        'RMSE': metrics['RMSE'],
        'MAE': metrics['MAE'],
        'R-squared': metrics['R-squared']
    })

df_metrics = pd.DataFrame(metrics_data)

# Sort models for better visualization, global models first, then by cluster
df_metrics['Sort_Key'] = df_metrics['Model'].apply(lambda x: 0 if 'Global' in x else int(x.split()[-1]))
df_metrics = df_metrics.sort_values(by='Sort_Key').drop(columns='Sort_Key')

print('Consolidated Metrics DataFrame:')
print(df_metrics)

# Plotting function for metrics
def plot_metrics_bar(df, metric_col, title, ylabel):
    plt.figure(figsize=(12, 6))
    bars = plt.bar(df['Model'], df[metric_col], color=['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan'])
    plt.title(title)
    plt.xlabel('Model')
    plt.ylabel(ylabel)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Add metric values on top of bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.001, round(yval, 4), ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

# Generate plots for each metric
plot_metrics_bar(df_metrics, 'RMSE', 'Model RMSE Comparison', 'RMSE')
plot_metrics_bar(df_metrics, 'MAE', 'Model MAE Comparison', 'MAE')
plot_metrics_bar(df_metrics, 'R-squared', 'Model R-squared Comparison', 'R-squared')
```
